```python
import numpy as np

def beetle_population(years, initial_first_year, initial_second_year):
    """
    Simulates beetle population dynamics over a number of years.

    Parameters:
        years (int): Number of years to simulate.
        initial_first_year (int): Initial population of first-year beetles.
        initial_second_year (int): Initial population of second-year beetles.

    Returns:
        first_year_population (list): Population of first-year beetles per year.
        second_year_population (list): Population of second-year beetles per year.
        third_year_population (list): Population of third-year beetles per year.
        total_population (list): Total beetle population per year.
    """

    # Transition matrix:
    # Row 0: Newborns from 2nd-year (2 offspring) and 3rd-year females (10 offspring, half female â†’ 5)
    # Row 1: Survival from year 1 to year 2 (30%)
    # Row 2: Survival from year 2 to year 3 (50%)
    A = np.array([
        [0, 1, 5],
        [0.3, 0, 0],
        [0, 0.5, 0]
    ])

    # Initial population vector: [year 1, year 2, year 3]
    population = np.array([initial_first_year, initial_second_year, 0])

    # Lists to store population data over time
    first_year_population = [population[0]]
    second_year_population = [population[1]]
    third_year_population = [population[2]]
    total_population = [np.sum(population)]

    # Simulate population changes over the specified number of years
    for _ in range(years):
        population = np.dot(A, population)
        first_year_population.append(population[0])
        second_year_population.append(population[1])
        third_year_population.append(population[2])
        total_population.append(np.sum(population))

    return first_year_population, second_year_population, third_year_population, total_population
```

```python
A = np.array([[0, 1, 5], [0.3, 0, 0], [0, 0.5, 0]])
n_t1 = np.array([[500], [500], [0]])
years = 150
for t in range(years):
    n_t1 = np.dot(A, n_t1)
    print(f"Year {t+1}: {n_t1.flatten()}")
# n_t1 is the population at time t
# A is the transition matrix
# n_t1 is the population at time t+1
# np.dot is the matrix multiplication function
# n_t1.flatten() is used to print the array in a single line
# years is the number of years to simulate
```

```python
# simulate for 150 years and print results
years = 150
first_year_pop, second_year_pop, third_year_pop, total_pop = beetle_population(years, 500, 500)

for i in range(len(total_pop)):
    print(f"Year {i}: First Year: {first_year_pop[i]:.2f}, Second Year: {second_year_pop[i]:.2f}, Third Year: {third_year_pop[i]:.2f}, Total: {total_pop[i]:.2f}")

```

```python
import matplotlib.pyplot as plt
# plot the results
plt.figure(figsize=(12, 8))
plt.plot(first_year_pop, label='First Year Beetles', color='blue')
ax = plt.gca()
# ensure axes start at 0 and give some headroom on y
ax.set_xlim(0, years)
ax.set_ylim(0, max(total_pop) * 1.1)
# place the spines so the axes cross at (0,0) and hide top/right spines
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# nicer x ticks (every 10 years)
plt.xticks(np.arange(0, years + 1, 10))
plt.plot(second_year_pop, label='Second Year Beetles', color='orange')
plt.plot(third_year_pop, label='Third Year Beetles', color='green')
plt.plot(total_pop, label='Total Beetle Population', color='red', linestyle='--')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Dynamics Over Time')
plt.legend()
plt.grid()
plt.show()
```
