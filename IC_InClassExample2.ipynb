```python
#import "population data.csv" into a pandas dataframe
import pandas as pd
```

```python
#Import "population data.csv" into a pandas dataframe
pop_data = pd.read_csv("population_Data.csv")
pop_data.head()
#View the dataframe
```

```python
# average the population from day 17 to day 60
avg_pop = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean()
print(f"Average population from day 17 to day 60: {avg_pop}")
```

```python
#plot the population data with "Days" on the x-axis and "Population" on the y-axis
import matplotlib.pyplot as plt
plt.plot(pop_data["Days"], pop_data["Population"], color='black', marker='o', linestyle='-')
plt.axhline(y=avg_pop, color='green', linestyle='--', label='carrying capacity')
plt.xlabel("Days")
plt.ylabel("Population")
plt.title("Population over Time")
plt.xlim(left=0)
plt.ylim(bottom=0)

# Add a separate y-axis label for the dashed line
ax = plt.gca()
ax2 = ax.twinx()
ax2.set_ylabel("y=276.5,carrying capacity", color='green')
ax2.set_yticks([avg_pop])
ax2.set_yticklabels([f'{avg_pop:.1f}'], color='green')
ax2.set_ylim(ax.get_ylim())

plt.legend(loc='upper left')
plt.show()
```
