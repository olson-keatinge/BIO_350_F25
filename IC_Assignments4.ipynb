```python
#Creating disease model using SIR model 
def SIR_model(S_0, I_0, R_0, beta, d_rate, t):
    # Save S, I, R values into lists
    S_Values = [S_0]
    I_Values = [I_0]
    R_Values = [R_0]
    # Set current S, I, R values
    current_S = S_0
    current_I = I_0
    current_R = R_0

    for _ in range(t):
        # Calculate new S, I, R values
        current_S = current_S - beta * current_S * current_I
        current_I = current_I - d_rate * current_I + beta * current_S * current_I
        current_R = current_R + d_rate * current_I

        # Round to three decimal places
        S_Values.append(round(current_S, 3))
        I_Values.append(round(current_I, 3))
        R_Values.append(round(current_R, 3))

    return S_Values, I_Values, R_Values

#call the SIR_model and give it S_0, I_0, R_0, beta, d_rate, t values and save them as 3 variables 
S_Values, I_Values, R_Values = SIR_model(999, 1, 0, 1/1000, 0.5, 30)

```

```python
#plot the population data with "Days" on the x-axis and "Population" on the y-axis
import matplotlib.pyplot as plt
plt.plot(S_Values, label='Susceptible', color='blue')
plt.plot(I_Values, label='Infected', color='red')
plt.plot(R_Values, label='Removed', color='gold')
plt.xlabel("Days")
plt.ylabel("Population")
plt.title("Fox Population over Time")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.grid()
plt.show()
```
