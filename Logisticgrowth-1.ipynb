```python
#Logistic Growth model, Density dependent models
#The logistic growth model is a common model used to describe the growth of populations. It is characterized by an initial exponential growth phase, followed by a slowing of growth as the population approaches its carrying capacity. The model is often represented by the equation:
#dN/dt = rN(1 - N/K)
#Where:
#N = population size
#r = intrinsic growth rate
#K = carrying capacity
#t = time
#Density dependent models take into account the effects of population density on growth and reproduction. In these models, the growth rate decreases as the population density increases, reflecting the limitations imposed by resources, space, and other factors. These models are important for understanding population dynamics and managing wildlife populations.

```

```python
def N_logistic_growth(N, s, d, K, time):
    #Define the variables
    N_Values = [N]
    current_N = N
    for _ in range(time):
        bN_t = s * current_N * (1 - current_N / K)
        dN_t = d * current_N
        current_N = current_N + bN_t - dN_t
        N_Values.append(current_N)
    return N_Values
```

```python
# Parameters
N_0 = 12      # initial population
s =  1    # germination rate
d=0.4     # death rate
K = 400       # carrying capacity
time = 35     # number of time steps

# Calculate population over time
result = N_logistic_growth(N_0, s, d, K, time)
print(result)
```

```python

days = list(range(time + 1))

# Plot the results with days on the x-axis and population on the y-axis
import matplotlib.pyplot as plt
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')
# Plotted the days versus results with marker o
plt.xlabel('Days')
# Added label for x axis
plt.ylabel('Plant population size')
# Added label for y axis
plt.title('Plant Population Growth Over Time')
# added title for the graph
plt.grid()
# added grid to graph for better readability
print(result)
```
