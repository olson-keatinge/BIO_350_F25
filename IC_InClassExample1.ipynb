```python
def N_exponential_growth(N_0, p, time):
    # Exponential growth model
    # Save N_values into a list
    N_Values = [N_0]
    # Save initial population
    current_N = N_0
    for t in range(time):
        current_N = current_N + p * current_N \
        # Round to three decimal places
        N_Values.append(current_N)
    return N_Values
```

```python
#Run the function with 40 days and p = 0.4 with an initial population of 0.01
result = N_exponential_growth(0.01, 0.4, 35)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis
import matplotlib.pyplot as plt
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')
# Plotted the days versus results with marker o
plt.xlabel('Days')
# Added label for x axis
plt.ylabel('Duckweed Population Size')
# Added label for y axis
plt.title('Duckweed Population Growth Over Time')
# added title for the graph
plt.grid()
# added grid to graph for better readablitiy 
plt.show()
```

```python
#change N_Exponential_growth to N_exp2 and have 3 conditions based on N_0
#if N_0<500, p=0.4
#if 500<=N_0<1000, p=0.2
#if N_0 >= 1000, p=0
def N_exp2(N_0, time):

    N_Values = [N_0]
    # Save initial population
    current_N = N_0
    for t in range(time):
        # Exponential growth model with conditionson p
        if current_N < 500:
            p = 0.4

        elif 500 <= current_N < 1000:
            p = 0.2
        else:
            p = 0
        current_N = current_N + p * current_N \
        # Round to three decimal places
        N_Values.append(current_N)
    return N_Values
```

```python
 #run the functionwith N_0= 0.01 and time = 35 
result = N_exp2(1, 35)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis
import matplotlib.pyplot as plt
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')
# Plotted the days versus results with marker o
plt.xlabel('Days')
# Added label for x axis
plt.ylabel('Population Size')
# Added label for y axis
plt.title('Exponential Growth of Population Over Time')
# added title for the graph
plt.grid()
# added grid to graph for better readablitiy 
plt.show()
```
